```python
import pandas as pd

# Load CSV files
df_main = pd.read_csv('instagram_cleaned.csv')
df_extra = pd.read_csv('insta.csv')

# Merge category column into main data
df_merged = df_main.merge(
    df_extra[['Channel Name', 'Main video category']],  # must be DataFrame
    on='Channel Name',                                  # common column
    how='left'
)

# Save final CSV
df_merged.to_csv('final_instagram_data.csv', index=False)
```
